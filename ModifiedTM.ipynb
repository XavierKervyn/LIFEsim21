```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

Definition of the range for delta

```python
delta = np.arange(0,1.1, 0.1)
```

Transmission function:

```python
def T3(X, Y, d):
    return (np.sin(X)**2 + d*(1 - np.sin(x)**2)) * (np.cos(Y)**2 + d*(1 - np.cos(Y)**2))
```

Create the (x,y) grid:

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.linspace(0, 2*np.pi, 100)

X, Y = np.meshgrid(x, y)

import decimal
# create an empty dictionary to store the transmission maps for each delta
path = 'C:/Users/kervy/Desktop/LIFEmission/LIFEsim/SavedFigures/'
for delta_ in delta:
    Z = T3(X, Y, delta_)
    numcontours = 30
    plt.contourf(X, Y, Z, 20, cmap='plasma', levels=np.linspace(0,1,numcontours))
    plt.colorbar()
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f"delta = {delta_.round(2)}")
    plt.savefig(path + f'TM_2Dillustrative_plot_delta{delta_}.png', dpi=300)
    plt.show()
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

delta = 0.2
x = np.linspace(-4, 4, 200)

def T(x):
    return np.sin(x)**2

def T_tilde(x):
    return np.sin(x)**2 * (1-delta) + delta

def T_hat(x):
    return np.sin(x)**2 * (1-delta*np.exp(-delta*x**2)) + delta*np.exp(-delta*x**2)

fig = plt.figure()
plt.plot(x, T(x), 'b')
plt.plot(x, T_tilde(x), 'r--')
# plt.plot(x, T_hat(x), 'r--')

leg = ['$\sin^2(x)$', '$\sin^2(x)(1-\delta) + \delta$' ] #, '$\sin^2(x)(1- \delta e^{-\delta x^2}) + \delta e^{-\delta x^2}$']
plt.legend(leg, loc='upper right', ncol=1)

plt.xlabel('x')
plt.ylim(0,1)
fig.set_size_inches(12, 4)
plt.savefig('C:/Users/kervy/Desktop/LIFEmission/LIFEsim/SavedFigures/' + 'TM_illustrative_plot.png', dpi=300)
plt.show()
```
